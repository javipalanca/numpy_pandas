# Ejercicio

Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos) total en los meses indicados.

```python
def balance(dataf, lista_meses):
    dataf["Balance"] = dataf["Ventas"] - dataf["Gastos"]
    df_meses = dataf[df.Mes.isin(lista_meses)]
    
    return df_meses[["Mes", "Balance"]]

balance(df, ["Enero", "Marzo"])
```
